## QUANTUM RANDOM NUMBER GENERATOR ~
---

The core of the quantum random number generator is to basically apply the hadamard gates to all the qubits(or wires). The Hadamard gate (H) is crucial here because it creates a superposition state. When applied to a qubit in the |0⟩ state, it puts the qubit into an equal superposition of |0⟩ and |1⟩. After the Hadamard gates, each qubit is in the state (|0⟩ + |1⟩)/√2. This means each qubit has an equal 50% chance of being measured as 0 or 1. When we measure these qubits in superposition, we get truly random outcomes due to the fundamental probabilistic nature of quantum mechanics.Each time we measure the pattern changes. Next is the classical processing part. This part involves interpreting and potentially transforming these binary outcomes into a usable random number which basically deals with the binaary to decimal conversion. The last step would be the Post-processing for uniformity, especially when there might be imperfections in the quantum hardware or potential biases in the raw output in order to enhance the quality of the random numbers produced.

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
from qiskit import transpile
```

```python

num_qubits= 29
qc=QuantumCircuit(num_qubits)
qc.h(range(num_qubits))
qc.measure_all()
qc.draw()
```

```python
# For execution
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
sim_result = simulator.run(compiled_circuit, shots=1024, memory=True).result()
counts = sim_result.get_counts()
print(counts)

rawvalues_sim = sim_result.get_memory()
print(rawvalues_sim)
```

```python
decimal_values = [int(bitstring, 2) for bitstring in rawvalues_sim]
print(decimal_values)
```

BROWNIAN MOTION SIMULATION USING QRNG:
---
Brownian motion is the random zig-zag motion of the particles in fluids. The particle's position will change over time due to collisions with surrounding fluid molecules. Each collision changes the particle's velocity in a random direction, and over time, this results in the particle exhibiting Brownian motion. Here I have used the QRNG to generate truly random numbers.The quantum-generated decimal values are mapped to angles (theta) and forces (F) to determine the direction and magnitude of movement for each simulation step.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = 0, 0
positions = [(x, y)]
# Number of steps in the simulation:
num_steps = len(decimal_values)  # using the number of quantum-generated values
delta_t = 1
```

```python
# Simulate Brownian motion using the quantum random numbers:
for i in range(num_steps):

    quantum_number = decimal_values[i]
    theta = (quantum_number % 360) * (2 * np.pi / 360)  # getting the angle in radians
    F = quantum_number / max(decimal_values)  # Normalizing the force to [0, 1]
    delta_x = F * np.cos(theta) * delta_t #calculating the change in position
    delta_y = F * np.sin(theta) * delta_t
    #updating the position:
    x += delta_x
    y += delta_y

    positions.append((x, y))

# Converting the positions to x and y arrays for plotting:
x_vals, y_vals = zip(*positions)

# Plotting the trajectory of the particle
plt.plot(x_vals, y_vals)
plt.title("Brownian Motion of a Particle Using QRNG")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.show()
```
